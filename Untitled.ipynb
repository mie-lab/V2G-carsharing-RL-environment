```python
 def update_reservation_state(self, reservations, reset):
    ## 3) timestamp of next reservation and 4) SOC needed for next trip ##

    # filter reservations: 
    # syscreatedate_daytime = 0: only reservations that were booked before t = 1
    # syscreatedate_daytime_endofday = 0: reservations that start at 00:00 of the next day (last time step before reset)
    if reset is True:
        time = 0
        next_reservation = np.ones(self.nr_vehicles) * -1
        duration_next_reservation = np.ones(self.nr_vehicles) * -1

    else:
        time = self.t + 1
        next_reservation = self.state[self.nr_vehicles * 2 :self.nr_vehicles *3]
        duration_next_reservation = self.state[self.nr_vehicles * 3 :self.nr_vehicles * 4]
        reservations_t0 =  reservations[reservations["syscreatedate_daytime"] == time]


    # iterate over all vehicles, search planned reservations
    count = 0
    for vehicle_id in self.vehicles_id:  
        # get reservation of car
        key = vehicle_id
        current_car = reservations_t0[reservations_t0["vehicle_no"] == vehicle_id]



        # remove reservations in past
        if len(self.planned_reservations_car[key]) != 0:
            if int(self.planned_reservations_car[key][0][0]) < self.t:
                self.planned_reservations_car[key] = self.planned_reservations_car[key][1:]

        # skip if no reservation found
        if len(current_car) == 0:
            if next_reservation[count] <  self.t and next_reservation[count] != -1:
                # save timestamp of next reservation
                if len(self.planned_reservations_car[key]) != 0:
                    next_reservation[count] = self.planned_reservations_car[key][0][0]

                    # save reservation duration
                    duration_next_reservation[count] = self.planned_reservations_car[key][0][1]

                    # remove reservation from dict
                    self.planned_reservations_car[key] = self.planned_reservations_car[key][1:]

                else:
                    next_reservation[count] = -1
                    duration_next_reservation[count] = -1
            count +=1

            continue

        for i in range(0,len(current_car)):    
            # get timestamp of next reservation, t = 96 if reservation starts at 00:00 of the next day
            try:
                reservation_of_car = current_car[current_car["reservationfrom_discrete"] != last_timestamp]
                reservation_time = reservation_of_car["reservationfrom_daytime"].iloc[i]
                reservation_duration = reservation_of_car["reservation_duration"].iloc[i]
            except:
                reservation_time = 96
                reservation_duration = current_car[current_car["reservationfrom_discrete"] == last_timestamp]["reservation_duration"].iloc[0]

            # if end of week, skipp reservations starting at 00:00 of next day (discrete tables only support reservations until 23:45 on the last day of a week)
            if end_of_week is True and reservation_time == 96:
                count +=1
                continue

            if i > 0 and reset is True:
                # save timestamp of next reservation
                # save reservation duration
                key = vehicle_id
                new_reservation = [reservation_time, reservation_duration]
                index = bisect.bisect_left([sublist[0] for sublist in  self.planned_reservations_car[key]], new_reservation[0])
                self.planned_reservations_car[key].insert(index, new_reservation)
            if reset is False: 
                new_reservation = [reservation_time, reservation_duration]
                index = bisect.bisect_left([sublist[0] for sublist in  self.planned_reservations_car[key]], new_reservation[0])
                self.planned_reservations_car[key].insert(index, new_reservation)

             # update only if reservation is before last planned reservation
        if self.planned_reservations_car[key][0][0] <  next_reservation[count]:
            # save timestamp of next reservation
            next_reservation[count] = self.planned_reservations_car[key][0][0]

            # save reservation duration
            duration_next_reservation[count] = self.planned_reservations_car[key][0][1]

            # remove reservation from dict
            self.planned_reservations_car[key] = self.planned_reservations_car[key][1:]
        count += 1
        
        return next_reservation, duration_next_reservation
    


```
